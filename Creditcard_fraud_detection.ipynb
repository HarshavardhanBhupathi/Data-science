93181 rows is still too large for SVM algorithm.lets downsize the data using resample from sklearn.utils

```python
cc1=cc[cc['Class']==1]
cc0=cc[cc['Class']==0]
```

```python
cc1.shape
```

```python
cc0.shape
```

```python
cc0_=resample(cc0,replace=False,n_samples=2000,random_state=42)
cc0_.shape
```

```python
data=pd.concat([cc0_,cc1])
data
```

```python
data.isnull().sum()
```

```python
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_train=s.fit_transform(x_train)
x_test=s.fit_transform(x_test)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
ypred
```

```python
accuracy_score(y_test,ypred)
```
